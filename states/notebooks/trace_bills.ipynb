Load text files into `data_dict`.

```python
text_files = os.listdir( data_path )
data_dict = {}
for tf in text_files:
    full_path = os.path.join( data_path, tf )
    with open( full_path ) as f:
        txt = f.read()
        
    data_dict[tf] = txt
```

Set the rolling n-gram size, and load a dictionary (`rolling_ngram_dict`) that contains lists of n-grams.

```python
ngram_size = 5
rolling_ngram_dict = {}
for tf in text_files:
    ngrams = []
    split_text = data_dict[tf].split()
    for i in range( 0, len( split_text ) - ngram_size + 1):
        ngrams.append( tuple( split_text[i:i+ngram_size] ) )
        
    rolling_ngram_dict[tf] = ngrams 
```

```python
def get_ngram_list( raw_text, ngram_size=5 ):
    ngrams = []
    split_text = raw_text.split()
    for i in range( 0, len( split_text ) - ngram_size + 1):
        ngrams.append( tuple( split_text[i:i+ngram_size] ) )
        
    return ngrams
```
